## Streaming

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

# 1. Load your GROQ_API_KEY from .env
load_dotenv()

# 2. Initialize the model using the correct Groq model ID
model = init_chat_model("qwen/qwen3-32b", model_provider="groq")

# 3. Invoke the model
response = model.invoke("write me a 200 words paragraph on Artificial Intelligence")
response
```

```python
for chunk in model.stream("write me a 200 words paragraph on Artificial Intelligence"):
    print(chunk.text , end ="|" , flush =True)
```

```python
for chunk in model.stream("why do parrots have colorful features"):
    print(chunk.text , end ="|" , flush =True)
```

### Batch
a collection of independent request to a model can improve performance and reduce costs , as the processing can be done in parallel.

```python
responses = model.batch([
    "why do parrots have colorful feathers?",
    "how do airplanes fly?",
    "what is quatum computing?",
])

for response in responses:
    print(response)
```

```python
responses = model.batch([
    "why do parrots have colorful feathers?",
    "how do airplanes fly?",
    "what is quatum computing?"],
    config = {
        'max_concurrency':5,
    }
)

for response in responses:
    print(response)
```
